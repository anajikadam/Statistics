## Definition of Geometric Mean
Geometric mean, also known as geometric average, is a way to calculate average for data that represent exponential or varying growth. It is defined as the product of the individual data points, all raised to the power of 1/n, where n is the sample size.

## Application of Geometric Mean
Geometric mean is typically used for data whose values are exponential in nature, or are supposed to be multiplied together.

1. Data which are exponential in nature
You can use geometric means to calculate things like:

Prevalence of diseases/Covid cases in a country, since some diseases grow exponentially, and having more cases now and will likely lead to having more cases in the future
Growth of a population
2. Data which are supposed to be multiplied together
In finance, geometric mean can be used to aggregate investment returns. For example, geometric mean is used to calculate CAGR (compound annual growth rate) which represents the constant rate of return over a given time period.
In statistics, geometric mean is used in the denominator of Bayesâ€™ rule between the likelihood and prior
One big caveat of geometric mean is that it only works with positive numbers, so you may need to consider a different method to calculate means if your data contains negative values.

The differences between arithmetic mean, geometric mean, and harmonic mean with simple examples:

1. **Arithmetic Mean**:
   - **Definition**: The arithmetic mean is the sum of all values in a dataset divided by the number of values.
   - **Formula**: Arithmetic Mean = sum_of_Datapoints/total_number_Datapoint,
   - **Example**: Consider the ages of four individuals: 20, 25, 30, and 35. The arithmetic mean age is ( 20 + 25 + 30 + 35)/4 = 27.5.

2. **Geometric Mean**:
   - **Definition**: The geometric mean is the \( n \)-th root of the product of \( n \) values.


3. **Harmonic Mean**:
   - **Definition**: The harmonic mean is the reciprocal of the arithmetic mean of the reciprocals of the values in the dataset.


**Key Differences**:

- **Arithmetic Mean**: Reflects the average value of a dataset and is influenced by extreme values.
- **Geometric Mean**: Used for averaging ratios, particularly when dealing with growth rates or ratios where relative changes matter.
- **Harmonic Mean**: Useful for averaging rates or speeds where the reciprocal of the values are averaged, ensuring that extreme values have less impact.

Each mean has its own specific application depending on the context of the dataset and what aspect of the data distribution or relationship you want to summarize.

# Variance


Variance is the spread of values in a dataset around its mean value. It tells you how far each number in the dataset is from its mean.

- Find the mean of the data
- Subtract the mean from each value to find the deviation from the mean
- Square the deviation from the mean
- Total the squares of the deviation from the mean for each value
- Divide by the degrees of freedom (sample size minus 1). This yields the unbiased estimator of the variance

# Covariance


covariance:

1. **Definition**: Covariance measures the directional relationship between two variables.

2. **Positive Covariance**: If covariance is positive, it means the two variables tend to increase or decrease together.

3. **Negative Covariance**: If covariance is negative, it means when one variable increases, the other tends to decrease.

4. **Example**: For example, you can use covariance to see the relationship between cats' weights and their body lengths.

5. **Calculation**: To calculate covariance, you can use the `cov` function from the numpy library in Python, which gives you a covariance matrix.

Here's a quick example in Python:
```python
import numpy as np

weights = [4, 5, 6, 7]
lengths = [30, 32, 34, 36]

cov_matrix = np.cov(weights, lengths)
print(cov_matrix)
```

This code will return a covariance matrix showing the covariance between weights and lengths of the cats.


image.png

The diagonal elements (in blue and green here) within the covariance matrix is the variance of each variable, while the other elements represent the covariance between the two variables. The covariance between cats' weight and body length is 3.925, a positive number, which means that the two variables tend to move in the same direction: the longer the cats' body is, the more they weigh.

If covariance of x and y is 0, then the two variables do not consistently move together. This does not mean that they are independent, just that they do not consistently move together.

image.png

Covariance â€” Problems with interpretation

1. **Interpretation Issue**: Covariance can be hard to interpret because it's affected by the scale of the data.

2. **Example**: For example, if you measure weights in grams (1000g) and kilograms (1kg), the covariance for grams will be higher.

3. **Scale Impact**: Larger values of x and y result in a larger covariance.

4. **Positive Covariance**: A positive value (e.g., 1000) indicates that x and y move in the same direction, but it doesn't show the strength of the relationship.

5. **Solution**: To better understand the strength of the relationship, we use the correlation coefficient, which adjusts for the scale of the data.

# Correlation


1. **Definition**: Correlation, like covariance, describes the directional relationship between two variables.

2. **Positive Correlation**: A positive correlation means the two variables tend to increase or decrease together.

3. **Negative Correlation**: A negative correlation means when one variable increases, the other tends to decrease.

4. **Formula**: The most common formula is Pearson's correlation, which is the covariance of two variables divided by the product of their standard deviations.

5. **Example Calculation**: Using the same data for cats' weights and lengths, you can calculate the correlation using the `corrcoef` function from the numpy library.

6. **Example Code**:
```python
import numpy as np
weights = [4, 5, 6, 7]
lengths = [30, 32, 34, 36]
correlation_matrix = np.corrcoef(weights, lengths)
print(correlation_matrix)
```

7. **Positive Correlation Result**: This code will give a positive correlation, e.g., 0.951, indicating a strong positive relationship.

8. **Diagonal Value**: The diagonal of the correlation matrix is always 1, representing the correlation of a variable with itself.


image.png

Unlike covariance which returns results in an unbounded range, the correlation coefficients return results in the range of [-1, 1] as it is essentially covariance normalized by standard deviation. This means that we can use correlation on datasets with different scales, such as the grams vs. kilograms example that we have mentioned above.

Rule of thumb for interpreting correlation coefficients:

1. **Perfect Correlation (Â±1)**:
   - **Near Â±1**: Perfect correlation.
   - Positive: As one variable increases, the other also increases.
   - Negative: As one variable increases, the other decreases.

2. **High Degree (Â±0.70 to Â±0.90)**:
   - Strong correlation.
   - Positive: High positive correlation.
   - Negative: High negative correlation.

3. **Moderate Degree (Â±0.50 to Â±0.70)**:
   - Medium correlation.
   - Positive: Moderate positive correlation.
   - Negative: Moderate negative correlation.

4. **Low Degree (Â±0.30 to Â±0.50)**:
   - Small correlation.
   - Positive: Low positive correlation.
   - Negative: Low negative correlation.

5. **Negligible Correlation (0.00 to Â±0.30)**:
   - Negligible or no correlation.

6. **No Correlation (0)**:
   - When the value is zero, there is no correlation.

This rule helps to quickly interpret the strength and direction of the relationship between two variables based on the correlation coefficient.

The "6" in the formula for Spearman's Rank Correlation Coefficient (Ï) is a scaling factor used to normalize the coefficient. Here's a brief explanation:

Spearman's Rank Correlation Coefficient (Ï) measures the strength and direction of the monotonic relationship between two variables. Monotonic relationships are those where one variable consistently increases or decreases as the other variable does.

The formula for Spearman's Ï between two sets of ranks \( R_1 \) and \( R_2 \) for \( n \) pairs of observations is:

\[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \]

where:
- \( d_i \) is the difference between the ranks of corresponding pairs of observations,
- \( n \) is the number of pairs of observations.

The "6" in the numerator (specifically in the denominator of the fraction) is a constant factor that standardizes the calculation of Spearman's Ï to ensure it falls within the range of -1 to 1, similar to Pearson's correlation coefficient. It adjusts the scale of the sum of squared rank differences to provide a standardized measure of the strength of monotonic association.

In summary, the "6" in Spearman's Rank Correlation Coefficient formula serves as a normalization factor to standardize the calculation of Ï, ensuring consistency and comparability of the coefficient across different datasets and sample sizes.

## Fixing Multicollinearity using VIF

Implementing VIF using Python
Now that you know how VIF is calculated, you can implement it using Python, with a little help from sklearn:

````python
import pandas as pd
from sklearn.linear_model import LinearRegression
def calculate_vif(df, features):
    vif, tolerance = {}, {}
    # all the features that you want to examine
    for feature in features:
        # extract all the other features you will regress against
        X = [f for f in features if f != feature]
        X, y = df[X], df[feature]
        # extract r-squared from the fit
        r2 = LinearRegression().fit(X, y).score(X, y)

        # calculate tolerance
        tolerance[feature] = 1 - r2
        # calculate VIF
        vif[feature] = 1/(tolerance[feature])
    # return VIF DataFrame
    return pd.DataFrame({'VIF': vif, 'Tolerance': tolerance})
````

# Calculating Chi2_score Ï‡2 manually

The first step is to create a contingency table.

````python
def chi2_by_hand(df, col1, col2):
    #---create the contingency table---
    df_cont = pd.crosstab(index = df[col1], columns = df[col2])
    display(df_cont)
    #---calculate degree of freedom---
    degree_f = (df_cont.shape[0]-1) * (df_cont.shape[1]-1)
    #---sum up the totals for row and columns---
    df_cont.loc[:,'Total']= df_cont.sum(axis=1)
    df_cont.loc['Total']= df_cont.sum()
    print('---Observed (O)---')
    display(df_cont)
    #---create the expected value dataframe---
    df_exp = df_cont.copy()
    df_exp.iloc[:,:] = np.multiply.outer(
        df_cont.sum(1).values,df_cont.sum().values)/df_cont.sum().sum()
    print('---Expected (E)---')
    display(df_exp)

    # calculate chi-square values
    df_chi2 = ((df_cont - df_exp)**2) / df_exp
    df_chi2.loc[:,'Total']= df_chi2.sum(axis=1)
    df_chi2.loc['Total']= df_chi2.sum()

    print('---Chi-Square---')
    display(df_chi2)
    #---get chi-square score---
    chi_square_score = df_chi2.iloc[:-1,:-1].sum().sum()

    return chi_square_score, degree_f
````

````python
chi_score, degree_f = chi2_by_hand(df,'Sex','Survived')
print(f'Chi2_score: {chi_score}, Degrees of freedom: {degree_f}')
````

For simplicity, we shall not go into the details of how to calculate the p-value by hand.

````python
def chi2_by_hand(df, col1, col2):
    #---create the contingency table---
    df_cont = pd.crosstab(index = df[col1], columns = df[col2])
    display(df_cont)
    ...
    chi_square_score = df_cont.iloc[:-1,:-1].sum().sum()

    #---calculate the p-value---
    from scipy import stats
    p = stats.distributions.chi2.sf(chi_square_score, degree_f)
    return chi_square_score, degree_f, p
````

## Defining the Hypotheses
You now define your null hypothesis and alternate hypothesis, just like the chi-square test. They are:

- Hâ‚€ (Null hypothesis) â€” that there is no difference among group means.
- Hâ‚ (Alternate hypothesis) â€” that at least one group differs significantly from the overall mean of the dependent variable.

We are now ready to begin our calculations for ANOVA. First, letâ€™s find the mean for each group:

Sum of squares of all observation â€” SS_total
Sum of squares within â€” SS_within
Sum of squares between â€” SS_between
Sum of squares of all observation â€” SS_total
The sum of squares of all observation is calculated by deducting each observation from the overall mean, and then summing all the squares of the differences:

## Using the statsmodels module to calculate f-score
Another way to calculate the f-value is to use the statsmodel module. You first build the model using the ols() function, and then call the fit() function on the instance of the model. Finally, you call the anova_lm() function on the fitted model and specify the type of ANOVA test to perform on it:

There are 3 types of ANOVA tests to perform,

````python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
df = pd.DataFrame(data, columns=['team','drink_type','reaction_time'])
model = ols('reaction_time ~ drink_type', data=df).fit()
sm.stats.anova_lm(model, typ=2)
````

The anova_lm() function also returns the p-value (0.983079). You can make use of the following rules to determine if the categorical variable has any influence on the numerical variable:

- if p < 0.05, this means that the categorical variable has significant influence on the numerical variable
- if p > 0.05, this means that the categorical variable has no significant influence on the numerical variable
- Since the p-value is now 0.983079 (>0.05), this means that the drink_type has no significant influence on the reaction_time.
